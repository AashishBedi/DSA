# Web Scraping- Grabbing a class

```python
res = requests.get('https://en.wikipedia.org/wiki/Nikola_Tesla')
```

```python
soup = bs4.BeautifulSoup(res.text, "lxml")
```

```python
soup
```

```python
soup.select('.templatequote')
```

```python
first_item = soup.select('.templatequote')[0]
```

```python
first_item
```

```python
first_item.text
```

```python
for i in soup.select('.templatequote'):
    print(i.text)
```
