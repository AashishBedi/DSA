```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Basic Graph
x = [5,10,15,20,25]
y = [6,12,18,24,30]
plt.plot(x,y)
plt.show()
```

```python
#For colouring the line
plt.plot(x,y,color = 'pink')
```

```python
#For marking in the graph
plt.plot(x,y,color = 'pink', marker = "*")
plt.show()
```

```python
plt.plot(x,y,color = 'pink', marker = "*", markeredgecolor = "black", markersize = 10)
plt.show()
```

```python
plt.plot(x,y,color = 'm', marker = "*",linestyle = '--', markeredgecolor = "black", markersize = 10)
plt.show()
```

```python
plt.plot(x,y,'r*--') #Shorter way for marking and coloring with linestyling
```

```python
#For heading of the graph
plt.title('Heading1')
```

```python
plt.plot(x, y, 'r*--')
plt.title('Heading1', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
```

```python
plt.plot(x,y,'r*--')
plt.xlabel('x-axis')
plt.ylabel('Y-axis')
#plt.title('Heading1', fontdict = {'fontname' : 'Comic Sans MS', 'fontSize' : 20})
plt.show()
```

```python
plt.plot(x,y,'b*--')
plt.legend(['Single element'])
plt.show()
```

```python
plt.plot(x,y,'b*--')
plt.savefig('Graph.png',dpi = 100) #savefig helps us to save graph
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mb = pd.read_csv("Countries.csv")

# Add the ".csv" extension to the file name
#D_compare = pd.read_csv(file_name)
plt.plot(mb.year, mb.india, 'b.-', label = 'India')
plt.plot(mb.year, mb.china, 'r.-', label = 'China')
plt.plot(mb.year, mb.japan, 'g.-', label = 'Japan')
plt.plot(mb.year, mb.america, 'm.-', label = 'America')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Mobile Price Year wise')

plt.legend()
plt.savefig('mobileprice.png', dpi = 300)
plt.show()
```

```python
#Histogram

bunch = [0,20,40,60]
x = [10,20,30,40]

plt.hist(x,bins = bunch, color = 'blue')
plt.xlabel('Bunch')
plt.ylabel('Data')
plt.legend(["ok"])
plt.show()
```

```python
#PieChart
y = np.array([20,30,35,40])
name = ["Ajay", "Vijay", "Ram", "Sham"]

plt.pie(y,labels = name)
plt.show()
```

```python
#Explode
namesexplode = [0,0,0,0.2]
plt.pie(y,labels = name, explode = namesexplode)
plt.show()
```

```python
#Shadow
#plt.pie(y,labels = name, explode = namesexplode, shadow = True, colors = my_colors)
plt.pie(y,labels = name, explode = namesexplode, shadow = True)
plt.savefig('pie.png', dpi = 100)
plt.show()
```

```python
#BoxPlot
test = np.random.normal(100,200,50)
plt.boxplot(test)
plt.show()
```

```python
#Multiple boxPlot
test = np.random.normal(100,200,50)
test1 = np.random.normal(10,70,50)
test2 = np.random.normal(200,30,40)
test3 = np.random.normal(40,20,80)
data = [test, test1, test2, test3]
plt.boxplot(data)
plt.show()
```
